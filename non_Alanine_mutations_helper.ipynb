```python
import pandas as pd
import numpy as np
```

```python
df_exp_data = pd.read_csv("single_mut_effects.csv")
```

```python
df_exp_data.head()
```

We have:

- 453 PRO -> P, Proline
- 498 HIS -> H, Histidine
- 502 ARG -> R, Arginine

```python
tmp_dict= {}
sims_done_arr= ["453P", "498H", "502R"]
for key, value in df_exp_data["mutation"].to_dict().items() :
    if value[1:] in sims_done_arr :
        tmp_dict[key] = df_exp_data.loc[key]
```

```python
results_df = pd.DataFrame.from_dict(tmp_dict, orient="index")
```

```python
results_df
```

```python
del_del_g_provided = results_df["bind_avg"].to_numpy()  
R = 1.98*1e-3 #1e-3 for cal to kcal conversion 
T =300  #room temp  
K_D_app = 3.9*1e-11 #from paper 
delta_g_deg_wild = -R * T * np.log(K_D_app)

tmp_K_mut = 10**(del_del_g_provided) * K_D_app
del_del_g_mut = -R * T * np.log(tmp_K_mut)

del_del_g_total =  -(del_del_g_mut- delta_g_deg_wild ) 

results_df["del_del_G"] = del_del_g_total
```
