```python
#index where we mut to A
df_idx_A = (df["mutant"] == "A")
df[df_idx_A]
```

```python
#the mutations we want to do : 
muts = df["mutation"][df_idx_A]
```

get all muts in the format we want

```python
all_muts= [] 
D_amino = ["V", "I", "L", "M" , "F", "Y", "W"]
for mut in muts : 
    if mut[0] in D_amino :
        mut_str = mut[0] +":" + mut[1:-1] +":" + mut[-1]
        all_muts.append(mut_str)
```
