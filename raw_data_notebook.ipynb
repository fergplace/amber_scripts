```python
#index where we mut to A
df_idx_A = (df["mutant"] == "A")
df[df_idx_A]
```

```python
#the mutations we want to do : 
muts = df["mutation"][df_idx_A]
```

get all muts in the format we want for simple amino acids: 

```python
simple_muts= [] 
D_amino = ["V", "I", "L", "M" , "F", "Y", "W"]
for mut in muts : 
    if mut[0] in D_amino :
        mut_str = mut[0] +":" + mut[1:-1] +":" + mut[-1]
        simple_muts.append(mut_str)
```

```python
start = 0
stop = 25
itts = math.ceil(len(all_muts)/25)
for i in range(0,itts):
    for j in range(start,stop) : 
        if (all_muts[j][0] != "A") and (all_muts[j][0] != "G"):
            print(f"{all_muts[j]}", end=" ")
    start = stop
    stop = min(stop+25, len(all_muts))
    print()
```
